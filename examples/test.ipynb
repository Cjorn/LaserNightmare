```python
import laserfun as lf

p = lf.Pulse(pulse_type='sech', fwhm_ps=0.05, epp=200e-12,
             center_wavelength_nm=1550, time_window_ps=7)

f = lf.Fiber(length=0.01, center_wl_nm=1550, dispersion=(-0.12, 0, 5e-6),
             gamma_W_m=1)

results = lf.NLSE(p, f, print_status=False)

results.plot()
```
